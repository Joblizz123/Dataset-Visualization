## Investigation Overview


> The overall goals of the presentation is to:
> 1. show the age distribution of the trips
> 2. show the trip durations

> 4. show the relationship between start day and trip durations with respect to the user type



## Dataset Overview

> This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from markupsafe import Markup

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('fordbike_clean.csv')
```

## (Visualization 1)



## What age range had the highest participation during the trips?

> **Age range: 20 - 40 years**

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,6))
bins = np.arange(0, df.member_age.max()+5, 5)
plt.hist(df.member_age, bins=bins, rwidth=0.95);
plt.title('Age Distribution of the riders')
plt.xlabel('Age of theriders [years]')
plt.ylabel('Frequency')
```



## (Visualization 2)



### Which user type dominated the ride?
 - **The Subscriber User type dominated the trips/rides**

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,6))
sns.countplot(data=df, x='member_gender', hue='user_type')
plt.xlabel('Gender of the Users')
plt.title('Gender Distribution by User types');
```

## (Visualization 3)



### What day did most of the bikes start?
 - Thursday

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,6))
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.countplot(data=df, x='start_day', order=day_order)
plt.title('Start day trips distribution')
```

## (Visualization 4)

### What is the relationship between start day and ride duration (in hours) by user type?
> **Customers are more active and involoved in the rides during the weekends than subscribers**

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,6))
sns.stripplot(data=df, x='start_day', y='duration_hour', hue='user_type', jitter=0.4, dodge=True, order=day_order)
plt.ylabel('Trips Duration [Hours]')
plt.xlabel('Trip Start Days')
plt.legend(loc=1)
plt.title('Effect of Trip Start day on Trip duration by the Users ')
```
